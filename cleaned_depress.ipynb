```python
# 우울 표현 사전 불러오기
dep_data = pd.read_csv("/content/drive/MyDrive/2023_연구학점/shared_dataset/csv/similarity80.csv")
```

```python
# 우울 표현 사전에서 빠져 있는 부분 채우기(dsm)
dsm_list = []
for s in dep_data['symptom']:
  if s == "우울기분":
    dsm_list.append(1)
  elif s == "흥미_즐거움감소":
    dsm_list.append(2)
  elif s == "체중변화":
    dsm_list.append(3)
  elif s == "불면_수면과다":
    dsm_list.append(4)
  elif s == "초조_지연":
    dsm_list.append(5)
  elif s == "피로_활력상실":
    dsm_list.append(6)
  elif s == "무가치감_및_죄책감":
    dsm_list.append(7)
  elif s == "사고력_집중력감소":
    dsm_list.append(8)
  elif s == "자살사고":
    dsm_list.append(9)
dep_data['dsm'] = dsm_list
```

```python
dep_data
```

```python
# 한국어 전처리 함수

def clean_korean_documents(documents):

    # 정제된 텍스트 담을 리스트
    clean_documents=[]

    # 텍스트 정제 (형태소 분석 및 어간 추출)
    for i, document in enumerate(documents): #documents는 list 형태
        clean_words = ''
        try:
          for word in okt.pos(document, stem=True): # 어간 추출
              if word[1] in ['Noun', 'Verb', 'Adjective', 'Adverb']: # 명사/동사/형용사/부사 만 추출
                clean_words += word[0]+','
          clean_words = clean_words[:len(clean_words)-1]
          clean_documents.append(clean_words)
        except:
          clean_documents.append(' ')

    return clean_documents
```

```python
word_list = dep_data['word_to'].tolist()
```

```python
cleaned_word = clean_korean_documents(word_list) # 전처리 결과
```

```python
dep_data["cleaned_word"] = cleaned_word
```
